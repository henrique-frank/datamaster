#Generative AI (Inteligência Artificial Generativa)

refere-se a uma categoria de técnicas de inteligência artificial que envolvem a criação de modelos capazes de gerar dados, como texto, imagens, áudio e outros tipos de conteúdo. Em vez de simplesmente classificar ou reconhecer dados existentes, modelos generativos têm a capacidade de criar novos dados que se assemelham a exemplos do conjunto de dados de treinamento.



As aplicações da Generative AI são diversas e incluem:

1. **Geração de Texto:** Criação de conteúdo, como artigos, histórias, relatórios e redação automática.

2. **Geração de Mídia:** Criação de imagens, arte, música e vídeo.

3. **Tradução Automática:** Tradução de texto de um idioma para outro.

4. **Geração de Conteúdo de Mídia Social:** Automação de postagens e respostas em redes sociais.

5. **Chatbots e Assistência Virtual:** Interações de conversação com sistemas automáticos de suporte ao cliente.

6. **Design de Produtos e Simulação:** Geração de designs, protótipos e simulações.

7. **Reconhecimento de Voz:** Conversão de fala em texto e vice-versa.

8. **Medicina e Pesquisa:** Geração de dados médicos sintéticos para pesquisa e treinamento de algoritmos.

Generative AI tem um impacto significativo em diversas indústrias e continua a evoluir, oferecendo oportunidades criativas e práticas em várias áreas.

Para obter uma chave de API da OpenAI, siga estas etapas:

1. Vá para o site da OpenAI: https://www.openai.com/

2. Procure informações sobre a API ou produtos relacionados da OpenAI.

3. Normalmente, você encontrará um botão ou link para "Get Started" (Começar) ou "Sign Up" (Inscrever-se). Clique nele.

4. Siga o processo de inscrição ou registro. Isso geralmente envolverá fornecer suas informações de contato, concordar com os termos de uso e, em alguns casos, aguardar a aprovação.

5. Após o registro, a OpenAI fornecerá a você uma chave de API exclusiva. Essa chave é essencial para acessar os recursos da API, como o GPT-3.

Lembre-se de que a disponibilidade da API da OpenAI pode variar de acordo com sua localização geográfica e a política da empresa, por isso, certifique-se de verificar se você está elegível para acessar a API. Certifique-se também de ler e entender os termos de uso da OpenAI antes de usar a API.

```python
with open("api_key.txt") as f:
    api_key = f.read()
```

```python
from openai import OpenAI

client = OpenAI(
  api_key=api_key,  # this is also the default, it can be omitted
)

completion = client.completions.create(
    model='text-embedding-ada-002',
    prompt= "Quero aprender mais sobre a API da OpenAI. Por favor, me ensine sobre suas principais funcionalidades e como posso usá-las em meu projeto."

    )

print(completion.choices[0].text)
print(dict(completion).get('usage'))
print(completion.model_dump_json(indent=2))
```
