# Declara√ß√µes `if`, `elif` e `else` ü§î

As declara√ß√µes `if` em Python nos permitem instruir o computador a executar a√ß√µes alternativas com base em um conjunto espec√≠fico de resultados.

Verbalmente, podemos imaginar que estamos dizendo ao computador:

"Se isso acontecer, execute alguma a√ß√£o."

Podemos ent√£o expandir a ideia ainda mais com as declara√ß√µes `elif` e `else`, que nos permitem dizer ao computador:

"Se isso acontecer, execute alguma a√ß√£o. Ou, se outra coisa acontecer, execute alguma outra a√ß√£o. Ou, se *nenhum* dos casos acima acontecer, execute esta a√ß√£o."

Vamos dar uma olhada na formata√ß√£o de sintaxe das declara√ß√µes `if` para ter uma ideia melhor disso:

```python

    executar a√ß√£o1

    executar a√ß√£o3
```

Essas declara√ß√µes condicionais nos permitem criar l√≥gica de ramifica√ß√£o em nossos programas, onde diferentes caminhos podem ser seguidos com base em condi√ß√µes espec√≠ficas. üõ§Ô∏èüîçüìù

# Loops `for` üîÑ

Um loop `for` atua como um iterador em Python; ele percorre os itens em uma *sequ√™ncia* ou qualquer outro objeto iter√°vel. Objetos que aprendemos que podem ser iterados incluem strings, listas, tuplas e at√© objetos iter√°veis de dicion√°rios, como chaves ou valores.

J√° vimos um pouco da declara√ß√£o `for` em se√ß√µes anteriores, mas agora vamos formalizar nossa compreens√£o.

Aqui est√° o formato geral de um loop `for` em Python:


Isso permite que voc√™ execute um conjunto de instru√ß√µes para cada item no objeto iter√°vel. üîÑüì¶üë©‚Äçüíªüë®‚Äçüíª

A declara√ß√£o `while` em Python √© uma das maneiras mais gerais de realizar itera√ß√£o. Uma instru√ß√£o `while` executar√° repetidamente uma √∫nica instru√ß√£o ou grupo de instru√ß√µes enquanto a condi√ß√£o for verdadeira. O motivo pelo qual √© chamada de 'loop' √© porque as instru√ß√µes de c√≥digo s√£o repetidas v√°rias vezes at√© que a condi√ß√£o n√£o seja mais atendida.

O formato geral de um loop `while` √©:

```
while teste:
    instru√ß√µes de c√≥digo
else:
    instru√ß√µes finais de c√≥digo
```

Vamos ver alguns exemplos simples de loops `while` em a√ß√£o. 

````python
x = 0

while x < 10:
    print('O valor de est√° atualmente em: ',x)
    print('x ainda √© menor que 10, adicionando 1 a x')
    x+=1
````

````python
print(x)
````

````python
10 < 10
````

````python
10 <= 10
````

Observe quantas vezes as instru√ß√µes de print ocorreram e como o loop `while` continuou at√© que a condi√ß√£o True fosse atendida, o que aconteceu quando x == 10. √â importante notar que, uma vez que isso ocorreu, o c√≥digo parou. Vamos ver como podemos adicionar uma declara√ß√£o `else`: 

````python
x = 0

while x < 10:
    print('O valor de est√° atualmente em: ',x)
    print('x ainda √© menor que 10, adicionando 1 a x')
    x+=1

else:
    print('Finalizado!')
````

Podemos usar as instru√ß√µes `break`, `continue` e `pass` em nossos loops para adicionar funcionalidades adicionais para v√°rios casos. As tr√™s instru√ß√µes s√£o definidas como:

- `break`: Sai do loop no qual est√°.
- `continue`: Vai para o topo do loop no qual est√°.
- `pass`: N√£o faz absolutamente nada. Pode ser utilizado para evitar levantamento de erro no c√≥digo.

Pensando nas instru√ß√µes `break` e `continue`, o formato geral do loop `while` se parece com isto:

```python
while teste:
    instru√ß√£o de c√≥digo
    if teste:
        break
    if teste:
        continue
else:
```

As instru√ß√µes `break` e `continue` podem aparecer em qualquer lugar dentro do corpo do loop, mas geralmente as colocamos mais aninhadas em conjunto com uma instru√ß√£o `if` para executar uma a√ß√£o com base em alguma condi√ß√£o.

Vamos dar uma olhada em alguns exemplos! 

````python
x = 0

while x < 10:
    print('x est√° atualmente em: ', x)
    print(' x ainda √© menor que 10, adicionando 1 a x')
    x += 1
    if x == 3:
        print('x == 3')
    else:
        print('continuando...')
        continue
````

Neste c√≥digo, estamos usando um loop while para iterar de 0 a 9. √Ä medida que x aumenta, imprimimos mensagens explicativas. Quando x atinge 3, exibimos uma mensagem especial ("x == 3") e, em seguida, usamos break para sair do loop imediatamente. Isso significa que o loop ser√° interrompido quando x atingir 3 e n√£o continuar√° a itera√ß√£o.

````python
x = 0

while x < 10:
    print('x est√° atualmente: ', x)
    print(' x ainda √© menor que 10, adicionando 1 a x')

    x = x + 1
    if x == 3:
        print('Interrompendo porque x == 3')
        break
    else:
        print('continuando...')
        continue

    
````

Observe como a outra instru√ß√£o **else** n√£o foi alcan√ßada e "continuando" nunca foi impresso!

Ap√≥s esses exemplos breves, mas simples, voc√™ deve se sentir confort√°vel usando as instru√ß√µes **while** em seu c√≥digo.

**Cuidado: √â poss√≠vel criar um loop infinito com as instru√ß√µes **while**. Por exemplo:**

````python
# N√£o execute este c√≥digo!!!!
while True:
    print("Estou preso em um loop infinito!")
````
